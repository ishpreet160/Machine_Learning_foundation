# Basic Graph

```python
import matplotlib.font_manager

font_files = matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
font_names = [matplotlib.font_manager.FontProperties(fname=font_file).get_name() for font_file in font_files]

print("Available fonts:")
for font_name in sorted(list(set(font_names))):
    print(f"- {font_name}")
```

[plot documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
from matplotlib.lines import lineStyles
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3),dpi=100)

# Line 1
#plt.plot(x,y,label='2x',color='pink',linewidth=2,marker=".",linestyle='--',markersize=10,markeredgecolor='blue')

# shorthand notation
#fmt='[color][marker][Line]'
plt.plot(x,y,'b*-',label='2x')

## Line 2

# select interval we want to plot points at
x2=np.arange(0,4.5,0.5)


# Plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')

# Plot remainder of graph as a dot
plt.plot(x2[4:],x2[4:]**2,'g--')

# Add a title (specify font parameters with fontdict)
plt.title('Our first Graph!',fontdict={'fontname':'Humor Sans','fontsize':20})

# X and Y labels
plt.xlabel('x axis')
plt.ylabel('y axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,7,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png',dpi=300)

# Show plot
plt.show()
```

**Bar Chart**

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)

patters=['/','o','*']
for bar in bars:
    bar.set_hatch(patters.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

# Real World Examples

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/gas_prices.csv')

plt.figure(figsize=(8,7))

plt.title('Gas Prices over Time (in USD)',fontdict={'fontweight':'bold','fontsize':18})

plt.plot(gas.Year,gas.USA,'b.-',label='USA')
plt.plot(gas.Year,gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
plt.plot(gas.Year,gas.Australia)


# Another Way to plot many values!

# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#   if country in countries_to_look_at:
#    plt.plot(gas.Year, gas[country], marker='.')

#for country in gas:
#  if country!='Year':
#   plt.plot(gas.Year,gas[country],marker='.',label=country)

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_prices.png',dpi=300)

plt.show()
```

# Histograms

```python
fifa=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/fifa_data.csv')
fifa.head()
```

```python
bins=[40,50,60,70,80,90,100]

plt.hist(fifa.Overall,bins=bins,color='#D7B4F3')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2022')

plt.savefig('histogram.png',dpi=300)

plt.show()
```

# Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

labels=['Left','Right']
colors=['#8D5EB7','#D7B4F3']

plt.pie([left,right],colors=colors,labels=labels,autopct='%.2f %%')

plt.title('Foot prefrences of FIFA Players')

plt.show()
```


# Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

# Box and Whiskers Chart

```python
plt.style.use('default')

plt.figure(figsize=(8,5), dpi=100)

barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']

labels=['FC Barcelona','Real Madrid','New England Revolution']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], tick_labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = 'pink' )
    # change hatch
    #box.set(hatch = '/')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('Overall Rating')

plt.show()
```
