# Filtering Data

Filtering data is essential for extracting relevant subsets based on conditions.

```python
bios=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv')

# Filter rows based on conditions
bios.loc[bios['height_cm']>210,['name','height_cm']]
bios[bios['height_cm']>210][['name','height_cm']]

# Multiple conditions
bios[(bios['height_cm']>215)&(bios['born_country']=='USA')]

# Filter by string conditions
bios[bios['name'].str.contains('Keith|Patrick',case=False)]

# Regex filters
bios[bios['born_date'].str.contains(r'^19',na=False)] #starting with 19
bios[bios['name'].str.contains(r'sen$',na=False)] #sen$ (ending with sen)
```

**NOTE: Regex expressions are very powerful**

use regex=False to remove regex search

```python
bios[bios['born_country'].isin(['USA'])& bios['name'].str.startswith('Kristin')]
```

```python
bios.query('born_country=="USA" and born_city=="Seattle"')
```

# Adding/Removing columns

Adding and removing columns is important for maintaining and analyzing relevant data.

```python
# Add a new column
coffee['price']=4.99
```

```python
import numpy as np

# Conditional column
coffee['new_price']=np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
```

```python
# Remove a column
coffee.drop(columns=['price'],inplace=True) #inplace=True for permanent change
```

```python
# Rename columns
coffee.rename(columns={'new_price':'price'},inplace=True)
```

```python
# Create new columns from existing ones
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
coffee
```

```python
bios_new=bios.copy()
bios_new['first_name']=bios['name'].str.split(' ').str.get(0)
bios_new['last_name']=bios['name'].str.split(' ').str.get(1)
bios_new
```

```python
bios_new['born_datetime']=pd.to_datetime(bios_new['born_date'])
```

```python
bios_new['born_year']=bios_new['born_datetime'].dt.year
bios_new[['name','born_year']]
```

```python
bios_new.to_csv('bios_new.csv',index=False) # Saved bios_new file as csv
```

```python
# lambda function
bios['height_category']=bios['height_cm'].apply(lambda x:'Short' if x<175 else ('Avg' if x<185 else 'Tall'))
bios.head()
```

```python
# any function you want
def cat(row):
  if row['height_cm']<175 and row['weight_kg']<70:
    return 'Lightweight'
  elif row['height_cm']<185 or row['weight_kg']<=80:
    return 'Middleweight'
  else:
    return 'Heavyweight'
bios['Category']=bios.apply(cat,axis=1)
bios.head()
```

# Handling Null Values

Handling null values is essential to ensure the integrity of data analysis.

```python
coffee.loc[[0,1],'Units Sold']=np.nan
```

```python
# Fill NaNs with a specific value
coffee=coffee.fillna(1000)
```

```python
coffee.fillna(coffee['Units Sold'].mean())
coffee.loc[[0,1],'Units Sold']=15

# Interpolate missing values
coffee.fillna(coffee['Units Sold'].interpolate()) #looks at the neighbours
```

```python
# Drop rows with NaNs
coffee.dropna(subset=['Units Sold'])
```

```python
coffee[coffee['Units Sold'].isna()] # is NaN
coffee[coffee['Units Sold'].notna()] # not NaN
```

# Aggregating Data

Aggregation functions like value counts and group by help in summarizing data efficiently.

```python
# Value counts
bios[bios['born_country']=='USA']['born_region'].value_counts().tail()
```

```python
# Group by and aggregation
coffee.groupby(['Coffee Type'])['Units Sold'].mean()
coffee.groupby(['Coffee Type'])['Units Sold'].sum()

coffee.groupby(['Coffee Type','Day']).agg({'Units Sold':'sum','price':'mean'})
```

```python
# Pivot table
pivot=coffee.pivot(columns='Coffee Type',index='Day',values='revenue')
pivot
```

```python
pivot.loc['Monday','Latte'] #count
```

```python
pivot.sum(axis=1)
```

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios['month_born']=bios['born_date'].dt.month
bios['year_born']=bios['born_date'].dt.year

bios.groupby([bios['year_born'],bios['month_born']])['name'].count().reset_index().sort_values('name',ascending=False)
#bios.groupby(bios['born_date'].dt.year)['name'].count()
#bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values('name',ascending=False)
```

# Advanced Functionality

Advanced functionalities such as rolling calculations, rankings, and shifts can provide deeper insights.

```python
# Shift
coffee['yesterday_revenue']=coffee['revenue'].shift(-2)
coffee
```

```python
coffee['pct_change']=coffee['revenue']/coffee['yesterday_revenue']
coffee
```

```python
# Rank
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values(['height_rank']).sample(10)[['name','height_rank']]
```

```python
#cumulative sum
coffee['cumsum']=coffee['revenue'].cumsum()
coffee
```

```python
# Rolling Window
latte=coffee[coffee['Coffee Type']=="Latte"].copy()
latte['3day']=latte['Units Sold'].rolling(3).sum()
latte
```
